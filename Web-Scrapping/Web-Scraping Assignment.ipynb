Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
def web_content(urlString):
    page =requests.get(urlString)
    soup = BeautifulSoup(page.content)
    return soup
page = web_content('https://en.wikipedia.org/wiki/Main_Page')
header = []
for i in page.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6']):
    header.append(i.text)
df= pd.DataFrame({"Header": header})
df
```

Write python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
def web_content(urlString):
    page =requests.get(urlString)
    soup = BeautifulSoup(page.content)
    return soup
page = web_content('https://presidentofindia.nic.in/former-presidents')
header = []
for i in page.find_all('h3'):
    header.append(i.text)
Term_of_office = []
for j in page.find_all('h5'):
    Term_of_office.append(j.text)
df1= pd.DataFrame({"Header": header ,"Term of office" : Term_of_office})
df1
```

Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
def web_content(urlString):
    page =requests.get(urlString)
    soup = BeautifulSoup(page.content)
    return soup
page = web_content('https://www.cnbc.com/world/?region=world')
news_headline = []
news_url = []
news_timestamp = []
for newItem in page.find_all('div', class_="LatestNews-headlineWrapper"):
    news_timestamp.append(newItem.find('time').text)
    news_headline.append(newItem.find('a', class_="LatestNews-headline").text)
    news_url.append(newItem.find('a', class_="LatestNews-headline")['href'])

df1= pd.DataFrame({"News Headline": news_headline ,"News Link" : news_url, "Time": news_timestamp})
df1
```

Write a python program to scrape the details of most downloaded articles from AI in last 90
days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details and make data frame

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
def web_content(urlString):
    page =requests.get(urlString)
    soup = BeautifulSoup(page.content)
    return soup
page = web_content('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
titles = []
authors = []
dates = []
urls = []

for article in page.find_all("li",class_="sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs"):
    titles.append(article.find('h2', class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg").text)
    urls.append(article.find('a', class_="sc-5smygv-0 fIXTHm")['href'])
    authors.append(article.find('span', class_="sc-1w3fpd7-0 dnCnAO").text)
    dates.append(article.find('span', class_="sc-1thf9ly-2 dvggWt").text)
df1= pd.DataFrame({"Paper Titles": titles ,"Paper URL" : urls, "Authors": authors, "Published Date": dates})
df1
```

Write a python program to scrape mentioned details from dineout.co.in and make data frame

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
def web_content(urlString):
    page =requests.get(urlString)
    soup = BeautifulSoup(page.content)
    return soup
page = web_content('https://www.dineout.co.in/delhi-restaurants/buffet-special')
rest_name = []
cuisine = []
Location = []
Ratings = []
Image_URL = []
for restaurant in page.find_all("div",class_="restnt-main-wrap clearfix"):
    cousineStr = (restaurant.find('span',class_="double-line-ellipsis").text).split("|")[1]
    rest_name.append(restaurant.find('a', class_="restnt-name ellipsis").text)
    cuisine.append(cousineStr)
    Location.append(restaurant.find('div', class_="restnt-loc ellipsis").text)
    Ratings.append(restaurant.find('div', class_="restnt-rating rating-4").text)
    Image_URL.append(restaurant.find('img', class_="no-img")['data-src'])                                  
df2= pd.DataFrame({"Restaurant Name": rest_name ,"Cuisine" : cuisine, "Location": Location, "Ratings": Ratings, "Image_URL": Image_URL})
df2
```
