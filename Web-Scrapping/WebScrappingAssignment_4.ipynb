1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank B) Name C) Artist D) Upload date E) Views

```python
driver=webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(5)

name=[]
try:
    names=driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        name.append(i.text)
except NoSuchElementException:
    name.append('No details available')
except StaleElementReferenceException:
    name.append('No details available')
    
time.sleep(2)    

rank=[]
try:
    ranks=driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranks:
        rank.append(i.text)
except NoSuchElementException:
    rank.append('No details available')
except StaleElementReferenceException:
    rank.append('No details available')

time.sleep(2)

artist=[]
try:
    artists=driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artists:
        artist.append(i.text)
except NoSuchElementException:
    artist.append('No details available')
except StaleElementReferenceException:
    artist.append('No details available')
    
time.sleep(2)    

views=[]
try:
    views_list=driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in views_list:
        views.append(i.text)
except NoSuchElementException:
    views.append('No details available')
except StaleElementReferenceException:
    views.append('No details available')
    
time.sleep(2)    

upload_date=[]
try:
    upload_date_list=driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload_date_list:
        upload_date.append(i.text)
except NoSuchElementException:
    upload_date.append('No details available')
except StaleElementReferenceException:
    upload_date.append('No details available')
    
time.sleep(2)    

df=pd.DataFrame({"Rank":rank,
                 "Name":name,
                "Uploader/artist":artist,
                "Date of upload":upload_date,
                "Views(in billion)":views})
df
```

2.Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Series
B) Place
C) Date
D) Time

```python
driver=webdriver.Chrome()
driver.get("https://www.bcci.tv/")
time.sleep(5)
optn = driver.find_element(By.XPATH, "/html/body/header/div[3]/div[2]/ul/div[1]/a[2]")
optn.click()
time.sleep(5)
Series = []
Place = []
Date = []
Time = []
international_fixture= driver.find_elements(By.ID, "match-card")
for international in international_fixture:
    series_elem = international.find_element(By.CLASS_NAME, "match-tournament-name")
    Series.append(series_elem.text)
    place_elem = international.find_element(By.CLASS_NAME, "match-place")
    Place.append(place_elem.text)
    date_elem = international.find_element(By.CLASS_NAME, "match-dates")
    Date.append(date_elem.text)
    time_elem = international.find_element(By.CLASS_NAME, "match-time")
    Time.append(time_elem.text)
    
df=pd.DataFrame({'Series':Series,
                'Place':Place,
                'Date':Date,
                'Time':Time,
                })
df
```

3. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details: A) Rank
B) State
C) GSDP(18-19)- at current prices
D) GSDP(19-20)- at current prices
E) Share(18-19)
F) GDP($ billion)

```python
driver=webdriver.Chrome() 
time.sleep(2)

driver.get("http://statisticstimes.com/")
driver.maximize_window()
time.sleep(2)

economy=driver.find_element(By.XPATH, "//div[@class='dropdown'][2]//div/a[3]")
try:
    economy.click()
except ElementNotInteractableException:
    driver.get(economy.get_attribute('href'))
time.sleep(5)

indian_states=driver.find_element(By.XPATH, "//a[contains(text(), 'GDP of Indian states')]")
indian_states.click()
time.sleep(5)


rank=[]
State=[]
GSDP_23=[]
GSDP_22=[]
share=[]
GDP_billion=[]
driver.execute_script("return document.getElementsByClassName('videoslider')[0].remove();");
stats_table_elements = driver.find_elements(By.XPATH, "//*[@id='table_id']/tbody/tr")
for stat_row in stats_table_elements:
    try: 
        stat_cell = stat_row.find_elements(By.TAG_NAME, "td")
        rank.append(stat_cell[0].text)
        State.append(stat_cell[1].text)
        GSDP_23.append(stat_cell[2].text)
        GSDP_22.append(stat_cell[3].text)
        share.append(stat_cell[4].text)
        GDP_billion.append(stat_cell[5].text)
    except NoSuchElementException:
        pass

df=pd.DataFrame({'Rank':rank,
                'State':State,
                'GSDP(22-23)':GSDP_23,
                'GSDP(21-22)':GSDP_22,
                'Share(22-22)':share,
                'GDP(Billion_$)':GDP_billion})
df
```

8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year

```python
driver=webdriver.Chrome() 
time.sleep(2)
driver.get("https://archive.ics.uci.edu/")
driver.maximize_window()
time.sleep(2)

btn = driver.find_element(By.LINK_TEXT, "View Datasets")
driver.execute_script("return arguments[0].scrollIntoView(true);", btn)
time.sleep(1)
btn.click()
time.sleep(1)   
#div.rounded-box.bg-base-100

urls = []
name=[]
for _ in range(0,10):
    data_sets = driver.find_elements(By.CSS_SELECTOR, "div.rounded-box.grid")
    for data in data_sets:
        desc = data.find_element(By.CSS_SELECTOR, "a.link-hover.link.text-xl.font-semibold")
        name.append(desc.text)
        urls.append(desc.get_attribute("href"))
    try: 
        netbtn = driver.find_element(By.CSS_SELECTOR, ".btn-group > button.btn-primary.btn-sm.btn:nth-child(2)")
        driver.execute_script("return arguments[0].scrollIntoView(true);", netbtn)
        netbtn.click()
        time.sleep(3)
    except ElementClickInterceptedException:
        driver.find_element(By.CSS_SELECTOR, ".btn-group > button.btn-primary.btn-sm.btn:nth-child(2)").click()

data_type=[]
default_task=[]
attr_type=[]
no_of_instances=[]
no_of_attributes=[]
year=[]   
for url in urls:
    driver.get(url)
    attributes = driver.find_elements(By.CSS_SELECTOR, ".col-span-4 > p.text-md")
    default_task.append(attributes[1].text)
    attr_type.append(attributes[3].text)
    data_type.append(attributes[2].text)
    no_of_instances.append(attributes[4].text)
    no_of_attributes.append(attributes[5].text)
    
df=pd.DataFrame({'Name':name,
                'Data_type':data_type,
                'Task':default_task,
                'Attribute_type':attr_type,
                'No_of_instances':no_of_instances,
                'No_of_attributes':no_of_attributes,
                })
df
```

5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the
following details:
A) Song name
B) Artist name
C) Last week rank
D) Peak rank
E) Weeks on board

```python
driver=webdriver.Chrome() 
browser=driver.get("https://www.billboard.com/")
driver.maximize_window()
time.sleep(2)


charts=driver.find_element(By.XPATH, "//li[@class='o-nav__list-item lrv-u-align-items-center lrv-u-flex']//a")
try:
    charts.click()
except ElementNotInteractableException:
    driver.get(charts.get_attribute('href'))
time.sleep(2)

hot_100=driver.find_element(By.XPATH, "//a[contains(text(), 'VIEW CHART')]")
driver.execute_script("return arguments[0].scrollIntoView(true);", hot_100)
try:
    hot_100.click()
except ElementClickInterceptedException:
    driver.get(hot_100.get_attribute('href'))
time.sleep(2)
song_name=[]
artist_name=[]
last_week_rank=[]
peak_rank=[]
weeks=[]
song_header = driver.find_elements(By.CSS_SELECTOR, ".lrv-a-unstyle-list > .lrv-u-width-100p > .lrv-a-unstyle-list")
for song in song_header:
    try:
        song_nm = song.find_element(By.CSS_SELECTOR, ".c-title")
        song_name.append(song_nm.text)
        song_prop = song.find_elements(By.CSS_SELECTOR, ".c-label")
        artist_name.append(song_prop[0].text)
        last_week_rank.append(song_prop[1].text)
        peak_rank.append(song_prop[2].text)
        weeks.append(song_prop[3].text)
    except NoSuchElementException:
        pass


df=pd.DataFrame({'Song_Name':song_name, 
                'artist_name':artist_name ,
                 'last_week_rank': last_week_rank,
                 'peak_rank': peak_rank,
                 'weeks':weeks        
                })
df
```

4. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
driver=webdriver.Chrome() 
browser=driver.get("https://github.com/")
driver.maximize_window()
opensource=driver.find_element(By.XPATH, "//button[contains(text(), 'Open Source')]")
opensource.click()
time.sleep(2)
trending=driver.find_element(By.XPATH, "//a[contains(text(), 'Trending')]")
trending.click()
time.sleep(2)

repo_urls=[]
Contributors_count=[]
Language_used=[]
repo_title=[]
repo_desc=[]

repos = driver.find_elements(By.CLASS_NAME, "Box-row")
for repo in repos:
    repo_title.append(repo.find_element(By.CSS_SELECTOR, ".h3.lh-condensed").text)
    try:
        repo_desc.append(repo.find_element(By.TAG_NAME, "P").text)
    except NoSuchElementException:
         repo_desc.append("-")
    Language_used.append(repo.find_element(By.XPATH, "//*[@itemprop='programmingLanguage']").text)
    Contributors_count.append(len(repo.find_elements(By.CSS_SELECTOR, "span.d-inline-block.mr-3 >a")))



df=pd.DataFrame({'Title':repo_title,
                'Description':repo_desc,
                'Contributors Count':Contributors_count,
                'Language Used':Language_used})
df
```

7. https://www.imdb.com/search/title/?title_type=tv_series&sort=num_votes,desc
1. Name 2.Year span 3.rating  4.votes

```python
driver=webdriver.Chrome() 
browser=driver.get("https://www.imdb.com/search/title/?title_type=tv_series&sort=num_votes,desc")
driver.maximize_window()
name = []
year_span= []
rating = []
votes = []

for _ in range(0,5):
    try:
        show_more_element = driver.find_element(By.CLASS_NAME, "ipc-see-more__text")
        driver.execute_script("return arguments[0].scrollIntoView(true);", show_more_element)
        show_more_element.click()
        time.sleep(2)
    except NoSuchElementException:
        pass
    except ElementClickInterceptedException:
        driver.execute_script("return arguments[0].scrollIntoView(true);", show_more_element)
        show_more_element.click()
        time.sleep(2)
        
header= driver.find_elements(By.XPATH, "//li[@class='ipc-metadata-list-summary-item']")

for tv_series in header:
    try:
   
        pattern=re.compile(r'[1-9]|[0-9]|[.]')
        title = tv_series.find_element(By.CLASS_NAME, "ipc-title__text")
        match= pattern.sub(' ', title.text)
        name.append(match)
        
        year = tv_series.find_element(By.CSS_SELECTOR, ".sc-be6f1408-8.fcCUPU.dli-title-metadata-item")
        year_span.append(year.text)
        
        rating_value = tv_series.find_element(By.CSS_SELECTOR, ".ipc-rating-star--imdb.ratingGroup--imdb-rating")
        rating.append(rating_value.text.replace('\n', ''))
   
        vote = tv_series.find_element(By.CSS_SELECTOR, ".sc-f24f1c5c-0.cPpOqU")
        votes.append(vote.text.replace('Votes', ''))
    except NoSuchElementException:
        pass


df=pd.DataFrame({'Name':name,
                'Years Span':year_span,
                'Rating':rating,
                'Votes':votes})
df
```
